#### Observations

- `Name (birth and death)` can be separated as `name`, `birth_year`, `death_year` columns
- `Term of office[23] Left office` has date + string data


```python
# split Name (birth and death) column into name and "Birth_and_death" columns

df[['Name', 'Birth_and_Death']] = df['Name (birth and death)'].str.split('\n', expand=True)
# display the dataframe
df
```

#### Observations

- `Birth_and_Death` can be separated as `birth_death` after removing unwanted string data


```python
# remove '(' , ')' and 'born ' text from the Birth_and_Death column
df['birth_death'] = df['Birth_and_Death'].map(lambda x: x.lstrip('(').rstrip(')').lstrip('born '))
```

```python
df
```

#### Observations

- `Name (birth and death)` and `Birth_and_Death` columns are not needed anymore as they have been separated in new column after cleaning the data.


```python
# drop column Birth_and_Death
df.drop(columns=['Birth_and_Death', 'Name (birth and death)' ], inplace = True)
df
```

#### Observations

- `birth_death` column can be separated as `birth_year`, `death_year` columns to get two columns of integer data.

```python
# see if you can get birth_year and death_year separated/
df[['birth_year', 'death_year']] = df['birth_death'].str.split('–', expand=True)
```

```python
df
```

#### Observations

- `birth_death`column can be dropped as we will not be using it further

```python
df.drop(columns=['birth_death'], inplace = True)
df
```

#### Observations

- We can convert `Term of office[23] Took office` into date object by splitting into `to_dd`,`to_mnthname` and `to_yyyy` columns

```python
df[['to_dd', 'to_mnthname', 'to_yyyy']] = df['Term of office[23] Took office'].str.split(' ', expand=True)
# display the dataframe
df
```

#### Observations

- We can convert `to_mnthname` into date object by adding another column .

```python
month_name = {'January' : 1, 'February' : 2, 'March': 3, 'April': 4, 
              'May': 5, 'June': 6, 'July': 7, 'August': 8, 'September': 9, 
              'October': 10, 'November': 11, 'December': 12 }
```

```python
df['to_month'] = df['to_mnthname'].map(lambda x: month_name[x]).apply(str)
df
```

#### Observations

- We can combine `to_dd`,`to_mnthname` and `to_yyyy` into one column.

```python
## convert Took Office into proper date column
df['took_office_date'] = df['to_yyyy'] +  '-' + df['to_mnthname'].map(lambda x: month_name[x]).astype(str) + '-' + df['to_dd']
df
```

```python
#To separate took office date column into 3 columns of integer values
df[['toy', 'tom','tod']] = df['took_office_date'].str.split('-', expand=True)
df
```

#### Observations

- `to_month`column is no longer needed and hence can be dropped

```python
df.drop(columns=['to_month'], inplace = True)
df
```

#### Observations

- We can add a new column named `total_days` to show the total tenure in days.

#### Observations

- We can create new columns `Name_new` and `total_days_new` and drop the duplicate values .

```python

dfc=df['Name']
dfd=dfc.drop_duplicates()
df['Name_new']=dfd
dfe=df['total_days']
dfg=dfe.drop_duplicates()
df['total_days_new']=dfg
df
```

```python
pm_frequency = (
    df.groupby(['Name'])['No.']
    .count()
    .reset_index()
)
pm_frequency.sort_values(by=['No.'], ascending=False, inplace=True)
```

```python
# Plotting play count distribution chart
s_df = pd.DataFrame(pm_frequency)

plt.figure(figsize=(15,8))
ax = sns.barplot(x = s_df['Name'],
            y = s_df['No.'].values,
            palette='Set1',
            data = s_df)
            # order=s_df.sort_values(, ascending=False).index)
plt.ylabel('# of times held PM') 
plt.xticks(rotation=45,
           horizontalalignment='right',
           fontweight='light',
           fontsize='medium' )
plt.title('# of terms', loc='center')
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset
cars =df1['Name_new']
 
data = df1['total_days_new']
 
 
# Creating explode data
explode = (0.1, 0.5, 0.1,0.2,0.2,0.4,0.2,0.4,0.2,0.2,0.2,0.1,0.4,0.2,0.2,)
 
# Creating color parameters
colors = ( "orange", "cyan", "fuchsia",
          "mediumseagreen", "lime", "hotpink",'teal','aquamarine','orange','red','olive',"lightpink","silver","khaki",'cornflowerblue')
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green", }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n".format(pct, absolute)
# Creating plot
fig, ax = plt.subplots(figsize =(20, 30))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = cars,
                                  shadow = False,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(fontsize="xx-large",color ="black"))
 
# Adding legend
ax.legend(wedges, cars,
          title ="PM",
          title_fontsize=20,
          loc ="best",
          fontsize=20,
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 15, weight ="bold")
ax.set_title("Percentage of days from total days all office has been Held",fontsize=25)
# show plot
plt.show()
```
